```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ["AI", "Cloud Computing", "Cybersecurity", "Blockchain", "IoT"]
values_2024 = [85, 70, 90, 60, 75]
values_2025 = [95, 80, 98, 72, 85]

x = np.arange(len(categories))  # locations
width = 0.4  # Width

# Creating the figure
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, values_2024, width, label="2024", color="#3498db")
bars2 = ax.bar(x + width/2, values_2025, width, label="2025", color="#e74c3c")

# Labels & title
ax.set_xlabel("Technology Sectors", fontsize=12, fontweight="bold")
ax.set_ylabel("Growth Index (%)", fontsize=12, fontweight="bold")
ax.set_title("Tech Trends Comparison 2025", fontsize=14, fontweight="bold", color="#2c3e50")
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=11)
ax.legend()

# Adding value
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}%', 
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha="center", fontsize=10, fontweight="bold")

# Credit
plt.figtext(0.9, 0.02, "Developed by Awais Manzoor", ha="right", fontsize=10, fontweight="bold", color="#2c3e50")

# Show the chart
plt.show()
```
