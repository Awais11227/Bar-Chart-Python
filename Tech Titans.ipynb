```python
import numpy as np
import matplotlib.pyplot as plt

# Revenue & Profit in billions
companies = ["Apple", "Microsoft", "Google", "Amazon", "Meta"]
revenues = [394, 211, 280, 524, 116]  # Hypothetical revenue values
profits = [99, 72, 76, 33, 39]  # Hypothetical profit values

# Bar 
x = np.arange(len(companies))
width = 0.4  

#  color scheme
colors = ["#007bff", "#28a745"]  # Blue for Revenue, Green for Profit

# stylish figure
fig, ax = plt.subplots(figsize=(12, 7), facecolor="#f5f5f5")  # Light background

# Plotting bars
bars1 = ax.bar(x - width/2, revenues, width, label="Revenue", color=colors[0], alpha=0.9, edgecolor="black")
bars2 = ax.bar(x + width/2, profits, width, label="Profit", color=colors[1], alpha=0.9, edgecolor="black")

# labels, title, 
ax.set_xlabel("Companies", fontsize=14, fontweight="bold", color="#333333")
ax.set_ylabel("Amount (Billion USD)", fontsize=14, fontweight="bold", color="#333333")
ax.set_title("💰 **Tech Titans: Revenue vs. Profit (Billion USD)** 🚀", fontsize=16, fontweight="bold", color="#222222", pad=15)
ax.set_xticks(x)
ax.set_xticklabels(companies, fontsize=12, fontweight="bold", color="#444444")
ax.legend(fontsize=12, loc="upper left", frameon=True, facecolor="white", edgecolor="black")


for bar in bars1 + bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, 
            f"{bar.get_height()}B", ha="center", fontsize=11, fontweight="bold", color="black")

# stylish grid
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#888888")
ax.spines["bottom"].set_color("#888888")

# Display the chart
plt.show()
```
