```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python


# Data
categories = ['A', 'B', 'C', 'D', 'E']
values_1 = [14, 46, 75, 32, 98]
values_2 = [56, 67, 12, 90, 35]

x = np.arange(len(categories))  # Positions for categories
width = 0.35  # Width of the bars

# Create figure
plt.figure(figsize=(9, 6), facecolor='whitesmoke')  # Background color added

# Plot grouped bar chart with enhanced styling
bars1 = plt.bar(x - width/2, values_1, width, label="Values 1", color='#4A90E2', edgecolor='black', alpha=0.9)
bars2 = plt.bar(x + width/2, values_2, width, label="Values 2", color='#FF8C42', edgecolor='black', alpha=0.9)

# Annotate values on bars
for bar in bars1 + bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2, f'{height}', ha='center', fontsize=10, fontweight='bold')

# Labels & Titles
plt.xlabel("Categories", fontsize=12, fontweight='bold', color='dimgray')
plt.ylabel("Values", fontsize=12, fontweight='bold', color='dimgray')
plt.title("Grouped Bar Chart\nDeveloped by Awais Manzoor", fontsize=14, fontweight='bold', color='darkblue', pad=15)

plt.xticks(ticks=x, labels=categories, fontsize=11)

# Legend & Grid
plt.legend(loc="upper left", fontsize=11, frameon=True, shadow=True)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show plot
plt.show()
```
