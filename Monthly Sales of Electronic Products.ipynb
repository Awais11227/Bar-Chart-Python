```python
import matplotlib.pyplot as plt
import numpy as np

# random sales data
products = ['Laptops', 'Smartphones', 'Tablets', 'Headphones', 'Smartwatches']
sales = np.random.randint(50, 500, size=len(products))  

# bar chart
plt.bar(products, sales, color='royalblue')

#  labels and title
plt.xlabel("Product Categories")
plt.ylabel("Units Sold")
plt.title("Monthly Sales of Electronic Products")

# Display values on top of bars
for i, v in enumerate(sales):
    plt.text(i, v + 10, str(v), ha='center', fontsize=10)

# Show the chart
plt.show()
```
